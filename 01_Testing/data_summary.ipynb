# Quick Explanation:

## The 'describe_df' function shows:
1) Shape of the Data Frame
2) Data Types
3) df.describe()

```python
def describe_df(df):
    print(f'''

Shape of Data Frame:
{df.shape}

DATA TYPES: 

{df.dtypes}

Name of Columns: 
{df.columns} 

COLUMNS DESCRIBED:

{df.describe()}
''')
```
